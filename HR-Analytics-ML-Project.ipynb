```python
top_players = df.groupby(
    "player_name"
)["runs_scored"].sum().nlargest(10)

top_players.plot(kind="bar")
plt.title("Top 10 Players by Runs")
plt.show()
```

```python
df["match_type"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.show()
```

```python
plt.hist(df["runs_scored"])
plt.title("Runs Distribution")
plt.show()
```

```python
country_runs = df.groupby(
    "country"
)["runs_scored"].sum()

country_runs.plot(kind="bar")
plt.show()
```

```python
plt.scatter(
    df["strike_rate"],
    df["runs_scored"]
)

plt.xlabel("Strike Rate")
plt.ylabel("Runs")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(
    x="role",
    y="wickets",
    data=df
)

plt.show()
```
